## 三、数据清洗与预处理

```python
# 数据清洗：转换日期格式、按日期排序
df['trade_date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')
df = df.sort_values('trade_date').reset_index(drop=True)

print("数据清洗完成！")
print("\n数据类型转换：")
print(df.dtypes)
print("\n按日期排序后前5条：")
print(df.head(3))
```

## 四、关键统计指标计算

```python
# 计算关键统计指标
start_price = df['close'].iloc[0]
end_price = df['close'].iloc[-1]
max_price = df['high'].max()
min_price = df['low'].min()
total_return = (end_price - start_price) / start_price * 100
amplitude = (max_price - min_price) / min_price * 100

avg_vol = df['vol'].mean()
avg_amount = df['amount'].mean()
total_vol = df['vol'].sum()
total_amount = df['amount'].sum()

up_days = (df['pct_chg'] > 0).sum()
down_days = (df['pct_chg'] < 0).sum()
flat_days = (df['pct_chg'] == 0).sum()
max_up = df['pct_chg'].max()
max_down = df['pct_chg'].min()

print("========== 关键统计指标 ==========")
print(f"\n【价格统计】")
print(f"  期初收盘价：¥{start_price:.2f}")
print(f"  期末收盘价：¥{end_price:.2f}")
print(f"  区间最高价：¥{max_price:.2f}")
print(f"  区间最低价：¥{min_price:.2f}")
print(f"  区间涨跌幅：{total_return:.2f}%")
print(f"  价格振幅：{amplitude:.2f}%")

print(f"\n【成交统计】")
print(f"  总成交量：{total_vol:,.0f} 手")
print(f"  日均成交量：{avg_vol:,.0f} 手")
print(f"  总成交额：{total_amount:,.0f} 千元")
print(f"  日均成交额：{avg_amount/1000:.2f}亿元")

print(f"\n【涨跌统计】")
print(f"  上涨天数：{up_days}")
print(f"  下跌天数：{down_days}")
print(f"  平盘天数：{flat_days}")
print(f"  最大单日涨幅：+{max_up:.2f}%")
print(f"  最大单日跌幅：{max_down:.2f}%")

print(f"\n【波动统计】")
print(f"  日均波动率：{df['pct_chg'].abs().mean():.2f}%")
print(f"  标准差：{df['close'].std():.2f}元")
```

## 六、可视化分析

### 6.1 图1：日K线图（含MA5/MA20均线）

> **交互说明：** 图表下方有滑动条，可拖拽缩放查看特定时间段；鼠标悬停显示当日开高低收数据。

```python
# 计算 MA5 和 MA20 均线
df['MA5'] = df['close'].rolling(window=5).mean()
df['MA20'] = df['close'].rolling(window=20).mean()

# 准备 K线图数据格式 [open, close, low, high]
kline_data = df[['open', 'close', 'low', 'high']].values.tolist()

# 准备日期标签
dates = df['trade_date'].dt.strftime('%Y-%m-%d').tolist()

print("K线图数据准备完成")
print(f"- 数据点数量：{len(kline_data)}")
print(f"- MA5 均线已计算")
print(f"- MA20 均线已计算")
print(f"\nK线图将在 HTML 中通过 ECharts 渲染展示")
```

**K线图解读：**
- 红色K线表示当日上涨（收盘 > 开盘），绿色K线表示当日下跌（收盘 < 开盘）
- MA5（橙色）反映短期趋势，MA20（蓝色）反映中期趋势
- 当MA5上穿MA20时形成"金叉"，视为潜在买入信号；反之形成"死叉"，视为潜在卖出信号

### 6.2 图2：每日收盘价走势曲线图

```python
# 收盘价曲线数据
close_prices = df['close'].tolist()
max_close = df['close'].max()
min_close = df['close'].min()
max_close_date = df.loc[df['close'].idxmax(), 'trade_date']
min_close_date = df.loc[df['close'].idxmin(), 'trade_date']

print("收盘价曲线数据准备完成")
print(f"- 最高收盘价：¥{max_close:.2f}（{max_close_date.strftime('%Y-%m-%d')}）")
print(f"- 最低收盘价：¥{min_close:.2f}（{min_close_date.strftime('%Y-%m-%d')}）")
print(f"- 收盘价标准差：{df['close'].std():.2f}元")
```
